## SDSS - dereddened

Only: `ra`, `dec`, `g`, `r`, `g_r` (deredenned using Schlegel maps).

The original query at (SDSS archive)[https://www.sdss.org/] was: `SELECT ra, dec, g, r from PhotoObjAll WHERE type = 6 and  clean = 1 and r>=10.0 and r<23.5;` (although split in small parts)

 * [162 million rows - 10GB](https://drive.google.com/file/d/0B8gjQokMGa4nUlRnZFlXSlZXOTQ/view?usp=sharing)

```python
ds = vaex.open("/Users/maartenbreddels/vaex/data/sdss/sdss_dereddened.hdf5")
```

```python
ds.healpix_plot(ds.col.healpix, show=True, f="log", healpix_max_level=9, healpix_level=9,
                healpix_input='galactic', healpix_output='galactic', rotation=(0,45)
               )
```
